```python
# To run this code you need to install the following dependencies:
# pip install google-genai

from google import genai
import os

def generate():
    client = genai.Client(api_key=os.environ.get("GEMINI_API_KEY"))

    result = client.models.generate_images(
        model="models/imagen-4.0-ultra-generate-preview-06-06",
        prompt="""YouTube thumbnail for a relaxing reggae music video mix: on the left side, featuring a beautiful woman with beach attire at the beach and a  title in reggae-insprired colours on the right side merging into a seamless picture together""",
        config=dict(
            number_of_images=1,
            output_mime_type="image/jpeg",
            person_generation="ALLOW_ADULT",
            aspect_ratio="16:9",
        ),
    )

    if not result.generated_images:
        print("No images generated.")
        return

    if len(result.generated_images) != 1:
        print("Number of images generated does not match the requested number.")

    for generated_image in result.generated_images:
        image = generated_image.image.show()


if __name__ == "__main__":
    generate()
```
